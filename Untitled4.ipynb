## 데이터 가공하기

```python
#시군구에서 시도 추출하여 새로운 행으로 만들기
apt2017['시도'] = apt2017.시군구.str.split(' ').str[0] 
apt2017['시도'].unique()
```

```python
apt2018['시도'] = apt2018.시군구.str.split(' ').str[0]
apt2018['시도'].unique()
```

```python
apt2019['시도'] = apt2019.시군구.str.split(' ').str[0]
apt2019['시도'].unique()
```

```python
#수도권, 비수도권 나누기
apt2017_cap = apt2017[(apt2017['시도'] == '경기도') | (apt2017['시도'] == '서울특별시') | (apt2017['시도'] == '인천광역시')]
apt2017_notcap = apt2017[(apt2017['시도'] != '경기도') & (apt2017['시도'] != '서울특별시') & (apt2017['시도'] != '인천광역시')]
```

```python
apt2018_cap = apt2018[(apt2018['시도'] == '경기도') | (apt2018['시도'] == '서울특별시') | (apt2018['시도'] == '인천광역시')]
apt2018_notcap = apt2018[(apt2018['시도'] != '경기도') & (apt2018['시도'] != '서울특별시') & (apt2018['시도'] != '인천광역시')]
```

```python
apt2019_cap = apt2019[(apt2019['시도'] == '경기도') | (apt2019['시도'] == '서울특별시') | (apt2019['시도'] == '인천광역시')]
apt2019_notcap = apt2019[(apt2019['시도'] != '경기도') & (apt2019['시도'] != '서울특별시') & (apt2019['시도'] != '인천광역시')]
```

```python
#3개년 데이터 통합하기
apt_3years = pd.concat([apt2017,apt2018,apt2019])
apt_cap_3y = pd.concat([apt2017_cap,apt2018_cap,apt2019_cap])
apt_notcap_3y = pd.concat([apt2017_notcap,apt2018_notcap,apt2019_notcap])
```

```python
#수도권, 비수도권 표현하는 변수 만들기
apt_3years['is_cap']=0
apt_3years.loc[(apt_3years['시도'] == '경기도') | (apt_3years['시도'] == '서울특별시') | (apt_3years['시도'] == '인천광역시') , 'is_cap'] = '수도권'
apt_3years.loc[(apt_3years['시도'] != '경기도') & (apt_3years['시도'] != '서울특별시') & (apt_3years['시도'] != '인천광역시') , 'is_cap'] = '비수도권'
```

수도권과 비수도권의 거래량이 비슷함을 확인할 수 있었다. 심지어 수도권의 거래량이 조금 더 많았다. 전국 17개의 시,도가 있고 그 중의 3개(서울특별시, 경기도, 인천광역시)만이 수도권이고, 나머지 14개의 시,도가 비수도권임을 감안하면 수도권 시,도의 평균 거래량이 훨씬 많은 것을 확인할 수 있었다.

-수도권/비수도권 면적당 거래금액 비교

```python
#수도권/비수도권 면적당 거래금액 비교
apt_3years["면적당거래금액"] = apt_3years["거래금액(만원)"] / apt_3years["전용면적(㎡)"] 
apt_cap_3y["면적당거래금액"] = apt_cap_3y["거래금액(만원)"] / apt_cap_3y["전용면적(㎡)"]
apt_notcap_3y["면적당거래금액"] = apt_notcap_3y["거래금액(만원)"] / apt_notcap_3y["전용면적(㎡)"] 
```

```python
sns.distplot(apt_cap_3y["면적당거래금액"],color="red",label="수도권")
sns.distplot(apt_notcap_3y["면적당거래금액"],color="green",label="비수도권")
plt.title("수도권/비수도권의 면적당 거래금액 분포")
plt.legend()
plt.xlabel('면적당 거래금액(만원)')
```

초록색이 비수도권의 그래프, 빨간색이 수도권의 그래프이다. 이를 보면 비수도권의 면적당 거래금액은 비교적 낮은 가격, 약 200만원에서 최빈값을 가지며, 주로 0-1000만원 사이에 분포한다. 수도권의 면적당 거래금액은 비교적 높은 금액, 약 500만원에서 최빈값을 가지며, 주로 0-1500만원 사이에 분포한다.

```python
#수도권/비수도권의 면적당 평균 거래금액 비교
sns.barplot(x="is_cap", y="면적당거래금액", data=apt_3years)
plt.title("수도권/비수도권 면적당 평균 거래금액 비교")
plt.xlabel('수도권/비수도권')
plt.ylabel('면적당 거래금액(만원)')
plt.show()
```

수도권의 면적당 거래금액 평균이 비수도권보다 약 2배 이상 높은 것을 확인할 수 있다.

```python
#수도권에서 가장 면적당 거래금액이 높은 매물 찾기
apt_cap_3y.loc[apt_cap_3y['면적당거래금액'].idxmax()]
```

경기도가 서울특별시보다 거래량이 약 1.5배가량 많았다. 

```python
#서울/경기 지역 거래금액 분포 비교
apt_Gyeongi_3y["면적당거래금액"] = apt_Gyeongi_3y["거래금액(만원)"] / apt_Gyeongi_3y["전용면적(㎡)"] 
apt_Seoul_3y["면적당거래금액"] = apt_Seoul_3y["거래금액(만원)"] / apt_Seoul_3y["전용면적(㎡)"]
apt_SG_3y["면적당거래금액"] = apt_SG_3y["거래금액(만원)"] / apt_SG_3y["전용면적(㎡)"] 
```

```python
sns.distplot(apt_Gyeongi_3y["면적당거래금액"],color="skyblue",label="경기")
sns.distplot(apt_Seoul_3y["면적당거래금액"],color="orange",label="서울")
plt.title("서울/경기의 면적당 거래금액 분포")
plt.legend()
plt.xlabel('면적당 거래금액(만원)')
```
